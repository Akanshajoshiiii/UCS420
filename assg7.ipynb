```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

roll_number = 102316015
np.random.seed(roll_number)
```

```python

sales_data = np.random.randint(1000, 5001, size=(12, 4))
sales_data
```

```python


months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']

sales_df = pd.DataFrame(sales_data, columns=categories, index=months)
sales_df
```

Part III: VisualizaƟons (Matplotlib and Seaborn)
1. Plot monthly sales trends for each category using line plots.
2. Create the following plots:
 Box plots to show the sales distribuƟon for each category.

```python
plt.figure(figsize=(12,6))
for category in categories:
    plt.plot(sales_df.index, sales_df[category], label=category)

plt.title('Monthly Sales Trend by Category')
plt.xlabel('Month')
plt.ylabel('Sales Units')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=sales_df[categories])
plt.title('Sales Distribution per Category')
plt.xlabel('Category')
plt.ylabel('Sales Units')
plt.show()
```

Q.2 For the array: array = np.array([[1, -2, 3],[-4, 5, -6]])
i. Find element-wise absolute value
ii. Find the 25th, 50th, and 75th percenƟle of flaƩened array, for each column, for each
row.
iii. Mean, Median and Standard DeviaƟon of flaƩened array, of each column, and
each row

```python
arr = np.array([[1, -2, 3], [-4, 5, -6]])
arr
```

```python
np.abs(arr)
```

```python

flat_arr = arr.flatten()


percentiles_flat = np.percentile(flat_arr, [25, 50, 75])
percentiles_flat
```

```python
# column
percentiles_col = np.percentile(arr, [25, 50, 75], axis=0)
percentiles_col

# row
percentiles_row = np.percentile(arr, [25, 50, 75], axis=1)
percentiles_row
```

```python

mean_flat = flat_arr.mean()
median_flat = np.median(flat_arr)
std_flat = flat_arr.std()

mean_flat, median_flat, std_flat
```

```python

mean_col = arr.mean(axis=0)
median_col = np.median(arr, axis=0)
std_col = arr.std(axis=0)

mean_col, median_col, std_col
```

```python

mean_row = arr.mean(axis=1)
median_row = np.median(arr, axis=1)
std_row = arr.std(axis=1)

mean_row, median_row, std_row
```
